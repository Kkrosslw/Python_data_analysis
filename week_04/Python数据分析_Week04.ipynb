# 数据框重塑
1. 用到的函数方法：drop,split,stack,merge

```python
df_创始人_准备 = df[["企业名称","掌门人/创始人"]]
df_创始人_准备
```

```python
# drop()
df_待合并 = df_创始人_准备.drop("掌门人/创始人",axis=1)  # axis表示轴
df_待合并.index.name = "序号"
df_待合并
# drop():删除某列
# axis=1 代表跨列，表示沿着每一行或者列标签模向执行对应的方法，也就是说将“掌门人/创始人”对应的列标签沿着水平方向依次删掉
# axis=0 代表往跨行，表示沿着每一列或行标签\索引值向下执行方法
```

```python
# split()
df["掌门人/创始人"].str.split(",",expand = True)
```

```python
df["掌门人/创始人"].str.split(",",expand = True).stack().head(20)
# stack()的意思是堆叠，而unstack()即不要堆叠
# stack() 将列索引变为行索引，且操作后返回对象是series类型，默认操作最里层数据
```

```python
# 以最内层为索引堆叠
df["掌门人/创始人"].str.split(",",expand = True).stack().reset_index(level=1,drop=True)
# reset_index 可以还原索引
# level控制了具体要还原的那个等级的索引，默认情况下移除所有级别
# drop ：在reset_index中默认为 drop = False ，drop为False则索引列会被还原为普通列，否则会丢失
```

```python
# 给新的Series命名
df_准备合并的列 =df['掌门人/创始人'].str.split(',', expand=True).stack().reset_index(level=1,drop=True).rename('掌门人/创始人')
df_准备合并的列.index.name = "序号"
df_准备合并的列
```

```python
dict_准备合并的列 = {"序号":df_准备合并的列.index,"掌门人/创始人":df_准备合并的列.values}
df_准备合并列=pd.DataFrame(dict_准备合并的列)
df_准备合并列.head(20)
```

# 数据框聚合
使用merge进行合并

```python
# 企业名称对应多个创始人
df_创始人 = pd.merge(df_待合并,df_准备合并的列,on="序号")
df_创始人
```

```python
# 整合代码
# 一个企业对应多个创始人
df_创始人_准备 = df[["企业名称","掌门人/创始人"]]
df_创始人_准备.index.name = "序号"
df_创始人 = pd.merge(df_创始人_准备.drop("掌门人/创始人",axis=1),\
                     df["掌门人/创始人"].str.split(",",expand=True) \
                                        .stack().reset_index(level=1,drop=True)
                                        .rename("掌门人/创始人"),on="序号")
display(df_创始人)


```

```python
# 整合代码
# 一个企业对应多个投资机构
df_机构_准备 = df[["企业名称","部分投资机构"]]
df_机构_准备.index.rename = "序号"
df_机构 = pd.merge(df_机构_准备.drop("部分投资机构",axis=1),\
                   df["部分投资机构"].str.split("[、,,]",expand=True) \
                                     .stack().reset_index(level=1,drop=True)
                                     .rename("部分投资机构"),on="序号")
display(df_机构)
```

# 练习
## 投资方对应多个企业

```python
df_投资方_企业 = df[["部分投资机构","企业名称"]]
```

```python
# 遍历所有的企业
# 先遍历每一行，再遍历每一行里的每一个
all_投资机构 = [item for sublist in df_投资方_企业["部分投资机构"].str.split("[、,,]") for item in sublist]
```

```python
# 尝试构建字典，投资方为key，企业名称为values
dict = {}
纪源资本 = df_投资方_企业[df_投资方_企业["部分投资机构"].str.contains("纪源资本")]
display(纪源资本)

# 存入字典
dict["纪源资本"] = 纪源资本["企业名称"].values
dict
```

```python
# 构建字典
org_dict = {}
for item in all_投资机构:
    df_投资_item = df_投资方_企业[df_投资方_企业["部分投资机构"].str.contains(item)]
    org_dict[item] = ",".join([i for i in df_投资_item["企业名称"].values])
org_dict
```

```python
# 转化为dataframe
投资方_独角兽企业_Series = pd.Series(org_dict)
display(投资方_独角兽企业_Series)
df_投资方_独角兽企业 = pd.DataFrame({"投资方":投资方_独角兽企业_Series.index,"独角兽企业":投资方_独角兽企业_Series.values})
display(df_投资方_独角兽企业)
```

```python
# 进行数据透视
df_投资方_独角兽企业.index.name = "序号"
df_投资方_独角兽企业_最终 = pd.merge(df_投资方_独角兽企业.drop("独角兽企业",axis=1),\
                            df_投资方_独角兽企业["独角兽企业"].str.split(",",expand=True).stack().reset_index(level=1,drop=True).rename("独角兽企业"),\
                            on="序号")
display(df_投资方_独角兽企业_最终)
```

```python
# 整合代码
df_投资方_企业 = df[["部分投资机构","企业名称"]]
# df_投资方_企业

# 遍历所有部分投资机构，为构建字典做准备
所有投资机构 = [item for sublist in df["部分投资机构"].str.split("[、,,]") for item in sublist]
# 所有投资机构

# 构建字典:所有投资方对应企业
org_dict = {}
for item in 所有投资机构:
    df_投资方_item = df_投资方_企业[df_投资方_企业["部分投资机构"].str.contains(item)]
    org_dict[item] = ",".join([i for i in df_投资方_item["企业名称"].values])
# org_dict

# 转化为dataframe
投资方_unicorn = pd.Series(org_dict)
df_投资方_unicorn = pd.DataFrame({"投资方":投资方_unicorn.index,"独角兽企业":投资方_unicorn.values})
# df_投资方_unicorn

# 最后一步：进行数据透视
df_投资方_unicorn.index.name = "序号"
df_投资方_unicorn_final = pd.merge(df_投资方_unicorn.drop("独角兽企业",axis=1),\
                               df_投资方_unicorn["独角兽企业"].str.split(",",expand=True).stack().reset_index(level=1,drop=True).rename("独角兽企业"),\
                               on="序号")
df_投资方_unicorn_final
```
